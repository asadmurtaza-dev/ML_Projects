```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("creditcard.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
print(df['Class'].value_counts())
print("Fraud percentage:", df['Class'].value_counts()[1]/len(df) * 100, "%")
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Class', data=df)
plt.title('Transaction Class Distribution')
plt.xticks([0,1], ['Normal','Fraud'])
plt.show()
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(df.corr(), cmap='coolwarm', annot=False)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df[df['Class']==0]['Amount'], bins=50, color='blue', label='Normal', alpha=0.6)
sns.histplot(df[df['Class']==1]['Amount'], bins=50, color='red', label='Fraud', alpha=0.6)
plt.title('Transaction Amount Distribution')
plt.legend()
plt.show()
```
